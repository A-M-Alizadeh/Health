```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution
data = np.random.normal(loc=75, scale=10, size=1000)  # Mean = 75, Std Dev = 10

# Define bins
bins = np.arange(40, 110, 5)  # Bin width = 5

# Plot histogram with density normalization
plt.hist(data, bins=bins, edgecolor='black', alpha=0.7, density=True)

# Plot the theoretical normal distribution curve
x = np.linspace(40, 110, 100)
pdf = (1 / (10 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - 75) / 10) ** 2)
plt.plot(x, pdf, color='red', linewidth=2, label="Gaussian PDF")

# Labels and title
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Histogram of Normally Distributed Data")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate Gaussian Data (mean=50, std=10)
gaussian_data = np.random.normal(loc=50, scale=10, size=100)

# Generate Exponential Data (=1, scale=10)
exponential_data = np.random.exponential(scale=10, size=100)

# Create Q-Q Plots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Q-Q Plot for Gaussian Data
stats.probplot(gaussian_data, dist="norm", plot=axes[0])
axes[0].set_title("Q-Q Plot of Gaussian Data")

# Q-Q Plot for Exponential Data
stats.probplot(exponential_data, dist="norm", plot=axes[1])
axes[1].set_title("Q-Q Plot of Exponential Data")

# Show the plots
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate Gaussian Data (Mean=50, Std=10)
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

# Compute Theoretical Quantiles
theoretical_quantiles = np.linspace(0, 1, len(data))  # Percentiles
expected_values = stats.norm.ppf(theoretical_quantiles, loc=50, scale=10)  # Normal quantiles

# Sort the actual data
sorted_data = np.sort(data)

# Plot Q-Q Plot
plt.figure(figsize=(6, 6))
plt.scatter(expected_values, sorted_data, label="Sample Data")
plt.plot(expected_values, expected_values, color="red", linestyle="--", label="Theoretical Line")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.title("Q-Q Plot Comparing Theoretical vs. Sample Quantiles")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate two Gaussian random variables
np.random.seed(42)
gaussian_data_1 = np.random.normal(loc=10, scale=2, size=100)   # N(10, 2^2)
gaussian_data_2 = np.random.normal(loc=10, scale=10, size=100)  # N(10, 10^2)

# Generate theoretical quantiles for standard normal N(0,1)
theoretical_quantiles = np.linspace(-3, 3, 100)  # Standard normal range
expected_values = stats.norm.ppf((np.arange(1, 101) - 0.5) / 100)  # Standard normal quantiles

# Sort the actual data
sorted_data_1 = np.sort(gaussian_data_1)
sorted_data_2 = np.sort(gaussian_data_2)

# Create Q-Q Plot
plt.figure(figsize=(7, 7))

# Plot the standard normal theoretical line (ideal N(0,1))
plt.plot(expected_values, expected_values, color='red', linestyle='--', label="N(0,1) Theoretical Line")

# Plot the measured quantiles for N(10, 2^2)
plt.scatter(expected_values, sorted_data_1, color='blue', alpha=0.6, label="N(10, 2^2)")

# Plot the measured quantiles for N(10, 10^2)
plt.scatter(expected_values, sorted_data_2, color='green', alpha=0.6, label="N(10, 10^2)")

# Labels and title
plt.xlabel("Theoretical Quantiles (Standard Normal)")
plt.ylabel("Measured Quantiles")
plt.title("Q-Q Plot: Comparing N(10, 2^2) and N(10, 10^2) to Standard Normal")

plt.legend()
plt.grid()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 10  # Number of samples per experiment
mu = 0  # Mean under H0
sigma = 1  # Standard deviation under H0
Ns = 10000  # Number of experiments

# Vector to store the t-statistics
tv = np.zeros(Ns)

# Perform simulations
np.random.seed(42)
for k in range(Ns):
    x = np.random.randn(N) * sigma + mu  # Generate samples from N(mu, sigma^2)
    m = np.mean(x)  # Sample mean
    s = np.std(x, ddof=1)  # Sample standard deviation (using N-1 for unbiased estimator)
    tv[k] = np.sqrt(N) * (m - mu) / s  # Compute t-statistic

# Sort absolute values of t-statistics
sorted_tv = np.sort(np.abs(tv))

# Compute empirical probabilities
probabilities = 1 - np.arange(Ns) / Ns

# Plot empirical survival function P(|T| > x)
plt.figure(figsize=(7, 5))
plt.plot(sorted_tv, probabilities, label="Empirical P(|T| > x)")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("P(|T| > x)")
plt.title("Empirical Survival Function of t-statistic")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm  # For normal distribution CDF

# Generate some sample data (e.g., from a normal distribution)
np.random.seed(42)  # for reproducibility
data = np.random.normal(0, 1, 1000)  # Mean 0, std dev 1, 1000 samples

# Calculate the CDF
x = np.sort(data)  # Sort the data
y = np.arange(len(x)) / len(x)  # Calculate CDF values

# Quantiles to visualize
quantiles = [0.25, 0.5, 0.75]  # Q1, Median, Q3
quantile_labels = ["Q1 (25th Percentile)", "Median (50th Percentile)", "Q3 (75th Percentile)"]


# --- Plotting ---

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="CDF")

# Plot the quantiles and their lines
for q, label in zip(quantiles, quantile_labels):
    # Interpolate to find the x value corresponding to the quantile
    x_q = np.interp(q, y, x)

    # Horizontal line
    plt.axhline(y=q, xmin=0, xmax=x_q, color='r', linestyle='--', linewidth=1)
    
    # Vertical line
    plt.axvline(x=x_q, ymin=0, ymax=q, color='r', linestyle='--', linewidth=1)

    # Add text label (adjust position as needed)
    plt.text(x_q + 0.1, q + 0.02, label, color='r')  # Adjust offset as needed


plt.xlabel("X (Sorted Data)")
plt.ylabel("F(x) (Cumulative Probability)")
plt.title("CDF with Quantiles")
plt.legend()
plt.grid(True)
plt.show()



# --- Alternative plotting using scipy.stats (if you know the distribution) ---
# This approach is more accurate if you know the underlying distribution

# x_range = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100) # For normal distribution
# plt.plot(x_range, norm.cdf(x_range), label="Normal CDF")

# # Plot quantiles (same logic as above) ...
# plt.show()
```
