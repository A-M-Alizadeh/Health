```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.metrics import accuracy_score

# Function to generate class labels based on the given formula
def generate_class(X1, X2):
    return np.sign(-2 * np.sign(X1) * np.abs(X1) ** (2/3) + 4 * X2 ** 2)

# Generate training data
Ntrain = 20000
X1_train = np.random.uniform(-1, 1, Ntrain)
X2_train = np.random.uniform(-1, 1, Ntrain)
C_train = generate_class(X1_train, X2_train)
X_train = np.column_stack((X1_train, X2_train))

# Plot training data
plt.figure(figsize=(6,6))
plt.scatter(X1_train[C_train == -1], X2_train[C_train == -1], color='red', label='Class -1')
plt.scatter(X1_train[C_train == 1], X2_train[C_train == 1], color='blue', label='Class 1')
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.title("Training Data")
plt.show()

# Train a decision tree classifier
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf.fit(X_train, C_train)

# Generate test data
Ntest = 100
X1_test = np.random.uniform(-1, 1, Ntest)
X2_test = np.random.uniform(-1, 1, Ntest)
C_test = generate_class(X1_test, X2_test)
X_test = np.column_stack((X1_test, X2_test))

# Predict using the trained classifier
C_test_pred = clf.predict(X_test)

# Plot test data with predictions
plt.figure(figsize=(6,6))
plt.scatter(X1_test[C_test_pred == -1], X2_test[C_test_pred == -1], color='red', label='Predicted -1', alpha=0.3)
plt.scatter(X1_test[C_test_pred == 1], X2_test[C_test_pred == 1], color='blue', label='Predicted 1', alpha=0.3)
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.title("Test Data Predictions")
plt.show()


# Export decision tree visualization
tree.export_graphviz(clf, out_file="Tree.dot", feature_names=['X1', 'X2'], 
                      class_names=['C=-1', 'C=1'], filled=True, rounded=True, special_characters=True)



# Generate a grid of points for plotting the decision boundary
grid_size = 50
x1_range = np.linspace(-1, 1, grid_size)
x2_range = np.linspace(-1, 1, grid_size)
X1_grid, X2_grid = np.meshgrid(x1_range, x2_range)
X_grid = np.column_stack((X1_grid.ravel(), X2_grid.ravel()))

# Predict class labels for the grid
C_grid_pred = clf.predict(X_grid).reshape(X1_grid.shape)

# Plot 3D Decision Boundary
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of the decision region
ax.plot_surface(X1_grid, X2_grid, C_grid_pred, cmap='viridis', edgecolor='k', alpha=0.7)

# Set labels and title
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("Predicted Class (C)")
ax.set_title("3D Decision Boundary of Decision Tree")

# Show the plot
plt.show()

# Compute accuracy
accuracy = accuracy_score(C_test, C_test_pred)
print("Accuracy:", accuracy)
```
